# Prototipo Experimental de Inferencia Local (Notebook Desordenado)
Este notebook contiene un borrador de código típico para realizar peticiones de prueba a un servidor Ollama.

```python
import requests
import json
```

```python
# Configuración mediante variables globales sueltas
MODELO = "llama3.2:3b"
URL = "http://localhost:11434/api/generate"
TEMP = "0.7"  # Cadena en lugar de float
MAX_TOKENS = 500
```

```python
prompt_usuario = "Explicar qué es la IA local"

# Petición HTTP directa sin try/except ni validación de esquema Pydantic
payload = {
    "model": MODELO,
    "prompt": prompt_usuario,
    "temperature": float(TEMP),
    "stream": False
}

# Sin control de excepciones de red ni verificación de estado HTTP
res = requests.post(URL, json=payload)
res_json = json.loads(res.text)

# Acceso directo a clave sin comprobación de tipos
print("Respuesta del modelo:")
print(res_json["response"])
```
